```python
%matplotlib inline
# Import libraries we will be using in this session
import control
from control import tf
import matplotlib
import numpy as np
import pylab as plt

matplotlib.rcParams['figure.figsize'] = (16.0,6.0)
matplotlib.rcParams.update({'font.size': 12, 'font.weight': 'bold'})
```

```python
# System parameters
zeta = [0,0.1,0.3,0.5,0.8,1.,2.] 
w_n = 0.1 #Hz
```

```python
#Step response to step with tf
fig, ax = plt.subplots(1,1)
lgd = []
for zeta_ in zeta:
    G = tf([w_n**2],[1, 2*zeta_*w_n, w_n**2])
    T = np.linspace(0,120,1000)
    T, y_step = control.step_response(G,T)
    ax.plot(T*w_n,y_step,lw=2-5)
    lgd += [str(zeta_)]
foo = ax.set_xlabel(r'$\omega _n t$')
foo = ax.set_ylabel(r'$c(t)$')
ax.set_xlim(0,12)
ax.set_ylim(-.02,2.02)
foo = plt.legend(lgd)
```

```python
#Step response to step with tf
fig, ax = plt.subplots(1,1)
lgd = []
for zeta_ in zeta:
    G = tf([w_n**2],[1, 2*zeta_*w_n, w_n**2])
    T = np.linspace(0,120,1000)
    T, y_step = control.impulse_response(G,T)
    ax.plot(T*w_n,y_step/w_n,lw=2-5)
    lgd += [str(zeta_)]
foo = ax.set_xlabel(r'$\omega _n t$')
foo = ax.set_ylabel(r'$\frac{c(t)}{w_n}$')
ax.set_xlim(0,12)
ax.set_ylim(-1.,1.)
foo = plt.legend(lgd)
```
