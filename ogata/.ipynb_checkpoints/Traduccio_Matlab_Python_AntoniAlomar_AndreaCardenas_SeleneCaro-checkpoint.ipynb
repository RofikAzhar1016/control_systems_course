```python
#MATLAB Program 5–6
t = np.linspace(0,10,100)
zeta = [0, 0.2, 0.4, 0.6, 0.8, 1]
n = 0
fig, ax = plt.subplots(1,1)
for zeta_ in zeta:
    num = [1]
    valor = zeta_ * 2
    den = [1, valor, 1]
    trans = tf(num,den)
    t, out = control.step_response(trans,t)
    plt.plot(t,out)

plt.grid(True)    
plt.title('Plot of Unit-Step Response Curves with $\omega_n = 1$ and $\zeta = 0, 0.2, 0.4, 0.6, 0.8, 1$')
plt.xlabel('t (sec)')
plt.ylabel('Response')
plt.text(4.1,1.86,'\zeta = 0')
plt.text(3.5,1.5,'0.2')
plt.text(3.5,1.24,'0.4')
plt.text(3.5,1.08,'0.6')
plt.text(3.5,0.95,'0.8')
plt.text(3.5,0.86,'1.0')

```

```python
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
import time
from __future__ import print_function

#MATLAB Program 5–6 CONTINUACIO
t = np.linspace(0,10,50)
zeta = [0, 0.2, 0.4, 0.6, 0.8, 1]
n = 0
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


X, Y = np.meshgrid(t, zeta)
out = []
for zeta_ in zeta:  
    num = [1]
    valor = zeta_ * 2
    den = [1, valor, 1]
    trans = tf(num,den)
    out_, t = control.step(trans,t)
    out += [out_]

wframe = ax.plot_wireframe(X, Y, np.array(out), rstride=1, cstride=1)

plt.title('Three-Dimensional Plot of Unit-Step Response Curves')
plt.xlabel('t Sec')
plt.ylabel('\zeta')
#plt.zlabel('Response')
```

```python
#MATLAB Program 5–7
num = [25]
den = [1, 6, 25]
t= np.linspace(0,5,1001)
trans = tf(num,den)
t, out = control.step_response(trans, t)
r = 1
while out[r] < 1.0001:
    r = r+1
rise_time = (r-1)*0.005

print("rise_time")
print (rise_time)

ymax = out.max(0)
i = 0
while i<1001:
    if (out[i] == ymax):
        break
    i = i+1
peak_time = i * 0.005

print("peak_time:")
print (peak_time)

max_overshoot = ymax-1
print("ymax:")
print(ymax)

s=1000
while out[s] > 0.98 and out[s] < 1.02:
    s = s-1

settling_time = (s-1) * 0.005
print("settling_time:")
print(settling_time)


plt.plot(t,out)
plt.grid(True)
```

```python
#MATLAB Program 5–13
t= np.linspace(0,12,120)
A = np.array([[-1,0.5],[-1,0]])
B = np.array([[0],[1]])
C = np.array([1,0])
D = 0
ss = control.ss(A,B,C,D)
Gcl = control.ss2tf(ss)
t,y_step = control.step_response(Gcl,t)

plt.plot(t,y_step, lw = 3)
plt.grid(True)
plt.title('Unit-Step Response')
plt.xlabel('t Sec')
plt.ylabel('Output')
```
