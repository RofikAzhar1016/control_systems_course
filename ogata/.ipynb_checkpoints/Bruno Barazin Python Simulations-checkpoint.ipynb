```python
%matplotlib inline
import pylab as plt
from control import tf
import control
import numpy as np
import matplotlib
matplotlib.rcParams.update({'font.size': 14})
matplotlib.rcParams['figure.figsize'] = (11.0, 6.)
```

```python
# Ejercicio 2.1
num1=[10]
den1=[1,2,10]

num2=[5]
den2=[1,5]

Hs=tf(control.series(tf(num1,den1),tf(num2,den2)))
print(Hs)

Hp=tf(control.parallel(tf(num1,den1),tf(num2,den2)))
print(Hp)

Hf=tf(control.feedback(tf(num1,den1),tf(num2,den2)))
print(Hf)
```

```python
#Ejercicios 5.1 y 5.2
A=np.array([[-1,-1],[6.5,0]])
B=np.array([[1,1],[1,0]])
C=np.array([[-1,0],[0,1]])
D=np.array([[0,0],[0,0]])
SpaceState=control.ss(A,B,C,D)
H=control.ss2tf(SpaceState)

T = np.linspace(0,10,1000)
T, y_step = control.step_response(SpaceState,T)

fig, ax = plt.subplots(2,2)
plt.sca(ax[0,0])
plt.plot(T,y_step[0])
plt.title("Step Response 1", weight = 'normal')
plt.grid(True)
plt.sca(ax[0,1])
plt.title("Step Response 2", weight = 'normal')
plt.plot(T,y_step[1])
plt.grid(True)
#En Matlab salen 4 gráficas, pero aquí solo obtenemos 2
```

```python
#Ejercicio 5.6

zeta = [0,0.2,0.6,0.8,1.] 
w_n = 1
fig, ax = plt.subplots(1,1)
lgd = []
for zeta_ in zeta:
    G = tf([w_n**2],[1, 2*zeta_*w_n, w_n**2])
    T = np.linspace(0,120,1000)
    T, y_step = control.step_response(G,T)
    plt.plot(T,y_step)
    lgd += [str(zeta_)]
foo = ax.set_xlabel(r'$t (sec)$')
foo = ax.set_ylabel(r'$Response$')
ax.set_xlim(0,19)
ax.set_ylim(-.05,2.05)
foo = plt.legend(lgd)
plt.title("Step Response Curves with w_n=1 and zeta=0,0.2,0.4,0.6,0.8,1", weight = 'normal')
plt.grid(True)
plt.show()
#Falta el plot3d!
#y_prime=np.diff(y_step)
```

```python
#Ejercicio 5.7

num=[25]
den=[1,6,25]

H=tf(num,den)

T = np.linspace(0,8,100)
T, y_step = control.step_response(H,T)
plt.plot(T,y_step)
plt.title("Step Response", weight = 'normal')
plt.grid(True)
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.show()
print(r'$t_p$={}'.format(T[np.argmax(y_step)]))
print(r'$M_p$={}'.format(np.max(y_step)-1))

#Faltan Rise Time y Settling Time
```

```python
#Ejercicio 5.13
A=np.array([[-1,0.5],[-1,0]])
B=np.array([[0],[1]])
C=[1,0]
D=[0]

T=np.linspace(0,12,120)
SpaceState=control.ss(A,B,C,D)
H=control.ss2tf(SpaceState)

T, y_step = control.step_response(SpaceState,T)
plt.plot(T,y_step)
plt.title("Unit-Step Response", weight = 'normal')
plt.grid(True)
plt.xlabel("Time (sec)")
plt.ylabel("Output")
plt.show()
```

```python
#Ejercicio 5.13 Exponencial
Uexp = np.exp(-T) #Exp
plt.plot(T,Uexp)

T, y_Uexp, foo = control.forced_response(H,T,Uexp)
plt.plot(T,y_Uexp)

plt.title("Response to Exponential Input u=e^-t", weight = 'normal')
plt.grid(True)
plt.xlabel("Time (sec)")
plt.ylabel("Exponential Input and System Output")
plt.show()
```
