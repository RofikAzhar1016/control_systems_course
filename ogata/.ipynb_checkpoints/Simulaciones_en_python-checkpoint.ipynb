```python
#Ejercicio 5-1
A1 = [[-1, -1], [6.5, 0]]
B1 = [[1, 1], [1, 0]]
C1 = [[1, 0], [0, 1]]
D1 = [[0, 0], [0, 0]]

T = np.linspace(0,25,1000)
tf = control.ss2tf(A1, B1, C1, D1)
T, step = control.step_response(tf, T, 0 , 0)
print(tf)
T, step2 = control.step_response(tf, T, 0 , 1)

plt.figure(1)
plt.plot(T, step[0], lw = 3)
plt.grid(True)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Amplitude", weight = 'bold')
plt.title('From: U1, To: Y1')

plt.figure(2)
plt.plot(T, step[1], lw = 3)
plt.grid(True)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Amplitude", weight = 'bold')
plt.title('From: U1, To: Y2')

plt.figure(3)
plt.plot(T, step2[0], lw = 3)
plt.grid(True)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Amplitude", weight = 'bold')
plt.title('From: U2, To: Y1')

plt.figure(4)
plt.plot(T, step2[1], lw = 3)
plt.grid(True)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Amplitude", weight = 'bold')
plt.title('From: U2, To: Y2')

```

```python
#Ejercicio 5-6

from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
import time
from __future__ import print_function

T = np.linspace(0,10,50)
zeta = [0, 0.2, 0.4, 0.6, 0.8, 1];

plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.title('Plot of Unit-Step Response Curves with \omega_n = 1 and \zeta = 0, 0.2, 0.4, 0.6, 0.8, 1')
plt.grid(True)

for n in zeta:
  num = [1]
  den = [1, 2*n, 1]
  tf_r = tf(num, den)
  T, step = control.step_response(tf_r, T)
  plt.plot(T, step,lw = 3)
  


n = 0
fig = plt.figure()
subplot = fig.add_subplot(111, projection='3d')


X, Y = np.meshgrid(T, zeta)
step = []
for n in zeta:  
    num = [1]
    den = [1, 2*n, 1]
    tf_r = tf(num, den)
    T, step_ = control.step_response(tf_r, T)
    step += [step_]

wframe = subplot.plot_wireframe(X, Y, np.array(step), rstride=1, cstride=1)

plt.title('Three-Dimensional Plot of Unit-Step Response Curves')
plt.xlabel('t Sec')
plt.ylabel('\zeta')
#plt.zlabel('Response')


```

```python
#Ejercicio 5-7

#Obtain the response of the system
num = [25]
den = [1,6, 25]
P_s = tf(num,den)
print(P_s)

T = np.linspace(0,5,1000)
T, y_step = control.step_response(P_s,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.grid(True)

#Check Rise time
tr = T[np.where(y_step>1)][0]
print('Rise Time: ' + repr(tr))

#Check Peak Time
tp = T[np.argmax(y_step)]
print('Peak Time: ' + repr(tp))

#Check Max Overshot
Mp = np.max(y_step)-1
print('Max Overshot: ' + repr(Mp))

#Check settling time
ts=0

for i in range(len(y_step) - 1,0, -1):
    if y_step[i] < 0.98 or y_step[i]>1.02:
        print('Settling Time: ' + repr(T[(i+1)]))
        break
```

```python
#Ejercicio 5-13
A = [[-1, 0.5], [-1, 0]]
B = [[0, 0], [1, 0]]
C = [[1, 0], [0, 0]]
D = [[0, 0], [0, 0]]

T = np.linspace(0,12,1000)
tf = control.ss2tf(A, B, C, D)
T, y = control.step_response(tf, T)


plt.figure(1)
plt.plot(T, y[0], lw = 3)
plt.xlabel("t Sec", weight = 'bold')
plt.ylabel("Output", weight = 'bold')
plt.title('Unit-Step Response')
plt.grid(True)


t = np.linspace(0,12,1000)
e = np.exp(-t)
u = [e, e]
u = np.transpose(u)
out = control.matlab.lsim(tf, u, t)

plt.figure(2)
plt.plot(t, out[0], lw = 3)
plt.plot(t, e, lw = 3)
plt.xlabel("t Sec", weight = 'bold')
plt.ylabel("Exponential Input and System Output", weight = 'bold')
plt.title('Response to Exponential Input u = exp(-t)')
plt.grid(True)

```
