```python
%matplotlib inline
import pylab as plt
from control import tf
import control
import numpy as np
import matplotlib
matplotlib.rcParams.update({'font.size': 18})

def plot_step_RLC(R,L,C):
    num = [0, 1]
    den = [L,R,1/C]
    tf_lcr = tf(num,den)
    print(tf_lcr)

    #Step response to step with tf
    T = np.linspace(0,25,1000)
    T, y_step = control.step_response(tf_lcr,T)
    plt.plot(T,y_step,lw = 3)
    plt.xlabel("Time [s]", weight = 'bold')
    plt.ylabel("Voltage [V]", weight = 'bold')
```

```python
# Exercise Undamped
# System parameters
R = 0 #Ohms
C = 1 #Farads
L = 1 #Henries

plot_step_RLC(R,L,C)
plt.grid(True)
```

```python
# Exercise underdamped
# System parameters
R = 1 #Ohms
C = 1 #Farads
L = 1 #Henries

plot_step_RLC(R,L,C)
plt.grid(True)
```

```python
# Exercise critically damped
# System parameters
R = 2 #Ohms
C = 1 #Farads
L = 1 #Henries

plot_step_RLC(R,L,C)
plt.grid(True)
```

```python
# Exercise overdamped
# System parameters
R = 10 #Ohms
C = 1 #Farads
L = 1 #Henries

plot_step_RLC(R,L,C)
plt.grid(True)
```

```python
#Why closed loops LCR Circuit inverse plant method
R = 0.5 #Ohms
C = 1 #Farads
L = 1 #Henries

num = [0, 1]
den = [L,R,1/C]
tf_lcr = tf(num,den)
tf_lcr_inv = tf(den,num)
print(tf_lcr)


#Step response to step with tf
vref = 1. * Ustep
T = np.linspace(0,25,1000)
T, y_step, foo = control.forced_response(control.series(tf_lcr_inv, tf_lcr),T,vref)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
plt.ylim([0,1.1])
```

```python
#Why closed loops LCR Circuit inverse plant method (Parametric uncertainty, 5% in capactitor magnitude)
R = 0.5 #Ohms
C = 1 #Farads
L = 1 #Henries

num = [0, 1]
den = [L,R,1/C]
tf_lcr = tf(num,den)
num_real = [0, 1]
den_real = [L,R,1/(1.05*C)]
tf_lcr_inv = tf(den_real,num_real)
print(tf_lcr)


#Step response to step with tf
vref = 3. * Ustep
T = np.linspace(0,25,1000)
T, y_step, foo = control.forced_response(control.series(tf_lcr_inv, tf_lcr),T,vref)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Voltage [V]", weight = 'bold')
```

```python
#Simple system, to test different control actions
matplotlib.rcParams['figure.figsize'] = (16.0,6.0)
T_sys = 1
P_s = tf([1],[T_sys, 1])
```

```python
#Original response
T = np.linspace(0,5,100)
T, y_step = control.step_response(P_s,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Value [units]", weight = 'bold')
plt.grid(True)
plt.ylim([0,1.2])
```

```python
#Proportional control action response
T = np.linspace(0,5,100)
K_s = 1
G_s_cl = control.feedback(control.series(K_s,P_s))
T, y_step = control.step_response(G_s_cl,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Value [units]", weight = 'bold')
plt.grid(True)
plt.ylim([0,1.2])
```

```python
#Integral control action response
T = np.linspace(0,10,100)
K_s = tf([1],[1, 0])
T_sys = 1
P_s = tf([1],[T_sys, 1])
G_s_cl = control.feedback(control.series(K_s,P_s))
T, y_step = control.step_response(G_s_cl,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Value [units]", weight = 'bold')
plt.grid(True)
plt.ylim([0,1.2])
```

```python
#Integral-Proportional control action response
T = np.linspace(0,10,100)
K_s = tf([1, 1],[0.10, 0])
G_s_cl = control.feedback(control.series(K_s,P_s))
T, y_step = control.step_response(G_s_cl,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Value [units]", weight = 'bold')
plt.grid(True)
plt.ylim([0,1.2])
```

```python
#Derivative control action response
T = np.linspace(0,10,100)
K_s = tf([0.1, 0],[1])
G_s_cl = control.feedback(control.series(K_s,P_s))
T, y_step = control.step_response(G_s_cl,T)
plt.plot(T,y_step,lw = 3)
plt.xlabel("Time [s]", weight = 'bold')
plt.ylabel("Value [units]", weight = 'bold')
plt.grid(True)
plt.ylim([0,1.2])
```
